# Introdução

Nosso dataset é relacionado a clientes de uma telefônica, desligados ou não, e nosso target é a coluna 'Churn' que diz se o cliente se desligou da empresa de telefônica ou não. **Aplicando dummies nas colunas com valores nao binarios**

```python
#Criando um dataset de clientes que abandonaram a empresa
customer_churn_yes = customer_churn.loc[customer_churn['Churn']=='Yes']
```
